```python
import cv2
import os

input_root = ""
output_root = ""

# Erstelle den Output-Ordner, falls er nicht existiert
if not os.path.exists(output_root):
    os.makedirs(output_root)

# Schleife durch alle Account-Ordner
for account in os.listdir(input_root):
    account_path = os.path.join(input_root, account)

    if not os.path.isdir(account_path):
        continue  # überspringe Dateien

    print(f"Bearbeite Account: {account}")

    # Erstelle Unterordner für jeden Account
    account_output = os.path.join(output_root, account)
    if not os.path.exists(account_output):
        os.makedirs(account_output)

    # Schleife durch alle Dateien im Account-Ordner
    for file_name in os.listdir(account_path):
        file_path = os.path.join(account_path, file_name)
        name, ext = os.path.splitext(file_name)
        ext = ext.lower()

        if ext not in [".mp4", ".mov", ".avi", ".mkv"]:
            continue

        cap = cv2.VideoCapture(file_path)
        ret, frame = cap.read()  # erstes Frame
        cap.release()

        if not ret:
            print(f"Fehler beim Lesen von {file_name}")
            continue

        # Frame speichern
        output_file = os.path.join(account_output, f"{name}_frame.jpg")
        cv2.imwrite(output_file, frame)

print("Fertig! Alle ersten Frames wurden extrahiert.")
```
